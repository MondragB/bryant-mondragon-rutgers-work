```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
danger_drop_speed = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
railgun_speed = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
time = np.arange(0,121, 10)
```

```python
danger_drop, = plt.plot(time, danger_drop_speed, color='red', label='Danger Drop')
railgun, = plt.plot(time, railgun_speed, color='green', label='RailGun')
plt.title('Coaster Speed Over Time')
plt.xlabel('Coaster Runtime')
plt.ylabel('Speed (MPH)')
plt.yticks(np.arange(min(time), max(time), 10))
plt.ylim(0,90)
plt.legend()
plt.grid()
```
