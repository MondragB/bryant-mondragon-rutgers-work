```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nj_fahrenheit = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
x_axis = np.arange(1,13,1)
```

```python
plt.plot(x_axis, nj_fahrenheit)
plt.show()
```

```python
nj_celsius = []
for month in nj_fahrenheit:
    nj_celsius.append((month - 32) * 0.56)
```

```python
plt.plot(x_axis, nj_celsius)
plt.show()
```

```python
plt.plot(x_axis, nj_fahrenheit)
plt.plot(x_axis, nj_celsius)
plt.show()
```

```python
plt.plot(x_axis, nj_fahrenheit, marker='+', color='blue',label="Fahrenheit")
plt.plot(x_axis, nj_celsius, marker='s', color='red', label="Celsius")
plt.legend(loc="upper right")
plt.show()
```
