```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
state_rain_df = pd.read_csv('../Resources/avg_rain_state.csv')
state_rain_df
states_only = state_rain_df['State'].tolist()
states_inches = state_rain_df['Inches'].tolist()

plt.bar(states_only, states_inches, color='salmon', align='edge')
plt.title('Average Rain Per State')
plt.xlabel('States')
plt.xticks(ticks=states_only, rotation='vertical')
plt.ylabel('Annual Rain (Per Inches)')
plt.ylim(0, max(states_inches)+5)
plt.tight_layout()
plt.savefig('testImg.png')

```
