```python
# Pandas
import pandas as pd

# SQL Alchemy
from sqlalchemy import create_engine
```

```python
# Create a connection to the Census database
database_path = '../Resources/Census_Data.sqlite'
engine = create_engine(f'sqlite:///{database_path}')

#create Connection
conn = engine.connect()
```

```python
# Query All Records in the the Census Table
data = pd.read_sql("SELECT * FROM Census_Data", conn)
data
```

```python
# Create a connection to the Zip database
database_path2 = '../Resources/zip_census.sqlite'
engine = create_engine(f'sqlite:///{database_path2}')

#create Connection
conn2 = engine.connect()
```

```python
# Query All Records in the Zip Table
data2 = pd.read_sql("SELECT * FROM Zip_Census", conn2)
data2
```

```python
# Merge the columns
combined_data = pd.merge(data, data2, on='CityState', how='inner')
```
