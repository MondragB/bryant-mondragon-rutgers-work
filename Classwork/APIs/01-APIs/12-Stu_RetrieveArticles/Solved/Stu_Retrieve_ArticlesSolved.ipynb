```python
# Dependencies
import requests
from config import api_key

url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"

# Store a search term
search_term = 'obama'

# Search for articles published between a begin and end date
begin_date= '20120101'
end_date= '20121231'

# Build url
query_url = f'{url}&api-key={api_key}&q={search_term}&begin_date={begin_date}&end_date{end_date}'
```

```python
# Retrieve articles
articles = requests.get(query_url).json()
articles['response']['docs']
```
