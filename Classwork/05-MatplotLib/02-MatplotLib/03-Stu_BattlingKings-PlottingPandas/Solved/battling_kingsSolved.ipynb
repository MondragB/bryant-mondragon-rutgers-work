```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
got_df = pd.read_csv('../Unsolved/Resources/got.csv')
got_df.head()
```

```python
# Get attacker and defender data
attacker_got_df = got_df['attacker_king'].value_counts()
defender_got_df = got_df['defender_king'].value_counts()
```

```python
# Get total battle data
battle_df = attacker_got_df.add(defender_got_df, fill_value=0)
```

```python
# Configure plot and ticks
battle_df.plot(kind='bar', facecolor='red')
plt.title("The Bloodthirst of Kings")
plt.ylabel("Number of Battles Participated In")
plt.xlabel('King')
```
