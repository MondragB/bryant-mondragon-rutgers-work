## 2 Features Into 1 Perceptron
Using 2 Features into 1 Perceptron in [TensorFlow Playground](http://playground.tensorflow.org) we obtain weights we can use to categorize the data of our model.

To determine the value returned by a perceptron first sum each product of the weights times their corresponding values and if that value is over a given threshold the perceptron returns a 1(signifying it's category).  If it is below the specified threshold it returns a 0.

```python
x_weight = -0.72
y_weight = -0.69
x_value = 5
y_value = 5
threshold = 0

if x_weight * x_value + y_weight * y_value <= threshold:
    #we indicate one categorization with the number 0
    output = 0
else:
    #we indicate the other categorization with the number 1
    output = 1

print(output)
```

We can create a function that will return the category of the values given their weights and a specified threshold.

```python
def perceptron_with_2_inputs(x_weight, y_weight, x_value, y_value, threshold):
    if x_weight * x_value + y_weight * y_value <= threshold:
        #we indicate one categorization with the number 0
        return 0
    else:
        #we indicate the other categorization with the number 1
        return 1

print(perceptron_with_2_inputs(
    x_weight = 0.74, 
    y_weight = 0.89, 
    x_value = 1, 
    y_value = 1, 
    threshold = 0))
```

We create a series of x and y values for the 2 features on our plot, then using the perceptron_with_2_inputs function we derive the categorization of those x, y values and use that to indicate the color on a scatter plot.

```python
x_y_values = [[x, y] for x in range(-6, 7) for y in range(-6, 7)]
x_values = [x_y_value[0] for x_y_value in x_y_values] 
y_values = [x_y_value[1] for x_y_value in x_y_values] 
colors = [perceptron_with_2_inputs(x_weight = 0.74, 
            y_weight = 0.89, 
            x_value = x_y_value[0], 
            y_value = x_y_value[1], 
            threshold = 0)
            for x_y_value in x_y_values]

plt.scatter(x_values, y_values, c=colors)
```

## 2 Features Into 2 Perceptrons
The model becomes significantly more complex with the addition of a single perceptron, though we can use our previous code as an outline.

```python
x_to_1_weight = 0.61
x_to_2_weight = -0.67
y_to_1_weight = 0.49
y_to_2_weight = -0.49
x_value = -3
y_value = -3
threshold = 0

perceptron_1_output = perceptron_with_2_inputs(
    x_weight = x_to_1_weight, 
    y_weight = y_to_1_weight, 
    x_value = x_value, 
    y_value = y_value, 
    threshold = 0)

perceptron_2_output = perceptron_with_2_inputs(
    x_weight = x_to_1_weight, 
    y_weight = y_to_1_weight, 
    x_value = x_value, 
    y_value = y_value, 
    threshold = 0)

final_output = perceptron_1_output and perceptron_2_output

print(final_output)
```

We can describe the behavior of our 2 perceptron model with a function.

```python
def two_perceptron_model(x_to_1_weight, x_to_2_weight, 
    y_to_1_weight, y_to_2_weight, x_value, y_value, threshold):
    perceptron_1_output = perceptron_with_2_inputs(
        x_weight = x_to_1_weight, 
        y_weight = y_to_1_weight, 
        x_value = x_value, 
        y_value = y_value, 
        threshold = 0)

    perceptron_2_output = perceptron_with_2_inputs(
        x_weight = x_to_1_weight, 
        y_weight = y_to_1_weight, 
        x_value = x_value, 
        y_value = y_value, 
        threshold = 0)

    return perceptron_1_output and perceptron_2_output
```

We can plot the behavior of our 2 perceptron model.

```python
x_y_values = [[x, y] for x in range(-6, 7) for y in range(-6, 7)]
x_values = [x_y_value[0] for x_y_value in x_y_values]
y_values = [x_y_value[1] for x_y_value in x_y_values]
colors = [two_perceptron_model(x_to_1_weight = 0.61, 
            x_to_2_weight = -0.67,
            y_to_1_weight = 0.49,
            y_to_2_weight = -0.49,
            x_value = x_y_value[0], 
            y_value = x_y_value[1], 
            threshold = 0)
            for x_y_value in x_y_values]

plt.scatter(x_values, y_values, c=colors)
```
